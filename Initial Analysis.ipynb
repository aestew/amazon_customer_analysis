```python
acb['est_purch_per_month'] = acb['Purchase_Frequency'].replace({
    'Few times a month': '3',
    'Less than once a month': '0.5',
    'Multiple times a week': '5',
    'Once a month': '1',
    'Once a week': '4'
})

acb['est_purch_per_month'] = pd.to_numeric(acb['est_purch_per_month'])

acb.head()
```

```python
labels = ['20s','30s','40s','50s']
acb['age_bin']=pd.cut(acb['age'],bins=range(20,70,10), labels = labels, right = False)
```

```python
pivot_acb_age_gender = acb.pivot_table(values = 'Timestamp', index = ['age_bin','Gender'],columns = 'est_purch_per_month',aggfunc = 'count')
pivot_acb_age_gender.sort_values(by = 'age_bin',ascending = True)
```

```python
pct_total_acb=pivot_acb_age_gender/len(acb)*100
pct_total_acb
```

```python
sns.heatmap(pct_total_acb, annot=True,cmap='YlGnBu')
```

```python
pivot_acb_gender = acb.pivot_table(values = 'Timestamp', index = ['Gender'],columns = 'est_purch_per_month',aggfunc = 'count')
pivot_acb_gender.sort_values(by = 'Gender',ascending = True)
pct_acb_gender=pivot_acb_gender/len(acb)*100
sns.heatmap(pct_acb_gender, annot=True,cmap='YlGnBu')
```
